```python
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
X,y = make_classification(n_samples=200,n_features=2,n_redundant=0,n_clusters_per_class = 2)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
model = Perceptron(max_iter=1000,eta0=0.1)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
X,y = make_classification(n_samples=300,n_features=4,n_informative=4,n_redundant=0,random_state=1)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
model = MLPClassifier(hidden_layer_sizes=(10,5) ,activation='relu',max_iter=500,learning_rate_init=0.01)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
X=np.array([[0,0],[1,0],[0,1],[1,1]])
y=np.array([[0],[1],[1],[0]])
np.random.seed(1)
weights = np.random.rand(2,1)
learning_rate=0.1
def sigmoid(x):
    return 1/(1+np.exp(-x))
def sigmoid_derivative(x):
    return x*(1-x)
for epoch in range(5000):
    output=sigmoid(np.dot(X,weights))
    error = y-output
    adjustments=learning_rate*np.dot(X.T,error*sigmoid_derivative(output))
    weights+=adjustments
print("Final Output: ",output)
```

```python
#Deep Learning
import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score
X,y = make_classification(n_samples=1000,n_features=4,n_informative=4,n_redundant=0,random_state=1)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
model = MLPClassifier(hidden_layer_sizes=(64,32,16) ,activation='relu',max_iter=500,learning_rate_init=0.01)
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print("Accuracy Score: ",accuracy_score(y_test,y_pred))
```

```python
import numpy as np
def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)
x_values = np.arange(-10, 11)

print("x\tReLU(x)\ttanh(x)")
for x in x_values:
    print(f"{x}\t{relu(x)}\t{tanh(x):.4f}")
```
