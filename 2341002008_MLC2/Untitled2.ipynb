```python
import numpy as np
x = np.array([0.5,0.2,0.1])
w = np.array([0.4,-0.7,0.2])
y_sum = np.dot(x,w)
output = y_sum
print("weight sum",y_sum)
print("output",output)
```

```python
def biological_neuron(signals,threshold=1.0):
    soma = sum(signals)
    if soma>=threshold:
        return 1
    return 0
signals = [0.3,0.4,0.5]
print("neuron output: ",biological_neuron(signals))
```

```python
def artificial_neuron(inputs,weights,bias=0):
    y_sum=np.dot(inputs,weights)+bias
    return y_sum
x = np.array([1,0,1])
w = np.array([0.5,-0.6,0.2])
print(artificial_neuron(x,w,bias=0.1))
```
