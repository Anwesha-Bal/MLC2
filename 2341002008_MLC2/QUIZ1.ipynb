```python
import numpy as np
import matplotlib.pyplot as plt
data = np.array([[10,20],[30,40],[15,19],[6,5],[20,25],[30,35],[48,20],[50,54],[11,16],[28,35]])
X = data[:,0]
Y = data[:,1]

plt.scatter(X,Y)
plt.title("QUIZ 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,6):
    m = KMeans(n_clusters=k,random_state=123)
    m.fit(data)
    wcss.append(m.inertia_)
plt.plot(range(1,6),wcss)
plt.xlabel("No. of clusters")
plt.ylabel("wcss")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3 , random_state = 123)
lable = kmeans.fit_predict(data)
plt.scatter(X,Y,c=lable)
plt.scatter(kmeans.cluster_centers_[0],kmeans.cluster_centers_[1],marker='x')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
def mcp_neuron(x1,x2,w1=-2,w2=-2,threshold=-3):
    y_sum = x1*w1+x2*w2
    return 1 if y_sum >=threshold else 0
for x1 in [0,1]:
    for x2 in [0,1]:
        print(x1,x2,"->",mcp_neuron(x1,x2))
```

```python
def Step(x):
    return 1 if x>=0 else 0
for i in [-0.5,1,0]:
    print(i,"->",Step(i))
```
