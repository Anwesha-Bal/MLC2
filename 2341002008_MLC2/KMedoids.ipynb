```python
import numpy as np
from sklearn_extra.cluster import KMedoids
import matplotlib.pyplot as plt
X=np.array([
    [1,2],[2,1],[1,1],[2,2],
    [6,7],[7,6],[6,6],[7,7],
    [3,8],[4,9]
])
kmedoids = KMedoids(n_clusters=3,random_state=42)
kmedoids.fit(X)
print("Mediods indices ",kmedoids.medoid_indices_)
print("mediods:")
print(kmediods.cluster_centers_)


print("Cluster labels:")
print(kmedoids.labels_)
plt.scatter(X[:,0],X[:,1],label='Data points')
plt.scatter(kmedoids.cluster_centers_[:,0],kmedoids.cluster_centers_[:,1],color='green')
plt.show()
```
